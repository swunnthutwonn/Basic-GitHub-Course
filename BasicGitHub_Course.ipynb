### Why we should learn terminology?

* essential to learn the commonly used terms
    * အသုံးများသော အသုံးအနှုန်းများကို သီးခြားလေ့လာသင့်သည်။
    <br/><br/>
* help understand the domain better and faster
    * မိမိလေ့လာသော အကြောင်းအရာကို ပိုမိုသိကျွမ်း နားလည်စေနိုင်သည်၊
    <br/><br/>
* understand with usage rather than dictionary interpretation
    * အဘိဓာန်ဖွင့်ဆိုမှုအတိုင်း မဟုတ်ဘဲ အသုံးပြုပုံကိုပါ တပါတည်းနားလည်နိုင်သည်။
    <br/><br/>
* understand with usage rather than direct translation into own language
    * တိုက်ရိုက်ဘာသာပြန်ခြင်းထက် အသုံးပေါ်မှုတည်၍ နားလည်နိုင်သည်။

### အသုံးများသော စကားလုံးများ (Terminology)

* Folder / Directory / URL
* Comand Prompt / Command Line / CLI
* (Remote) Repository - a folder or a place where your files are stored.
* Git - Verison Control Software
* GitHub - web hosting for Git Repository
* Staging area - temporary location
* Watch - get notification about a repository whenever there are changes
* Star - bookmark repository, easy access

### အသုံးများသော စကားလုံးများ (Git Commands)

* git init [repository_name]
    * <b>Repository</b> တစ်ခု စတင်တည်ဆောက်ရန် အသုံးပြုသည်
<br/><br/>
* git clone [url]
    * အွန်လိုင်းပေါ်ရှိ Repository အားမိမိ၏ ကွန်ပျူတာပေါ်သို့ ကူးယူရန် အသုံးပြုသည်။
<br/><br/>
* git add [file]
    * ဖိုင်တစ်ခု၏ ပြောင်းလဲမှုအား မှတ်မည်ဟုစေခိုင်းကာ Staging area တွင်ထားသည်။
<br/><br/>
* git add *
    * ဖိုင်များအားလုံး၏ ပြောင်းလဲမှုအား မှတ်မည်ဟုစေခိုင်းကာ Staging area တွင်ထားသည်။
<br/><br/>
* git reset [file]
    * Staging area တွင်ထားရှိပြီးသော ဖိုင်အား ပြန်လည်ရုပ်သိမ်းသည်။
    <br/><br/>
* git stauts
    * အတည်ပြုရန်လိုအပ်သော (ပြောင်းလဲမှုပြုထားသော) ဖိုင်များအား ဖော်ပြသည်။
    <br/><br/>
* git commit -m "[any_message_about_changes]"
    * ပြောင်းလဲမှုပြုလုပ်ထားသော ဖိုင်များအား အတည်ပြုမှတ်သားသည်။
    <br/><br/>
* git commit -a
    * add လုပ်ထားသော ဖိုင်များနှင့် ပြောင်းလဲမှုပြုထားသော ဖိုင်များအား အတည်ပြုမှတ်သားသည်။
    <br/><br/>
* git push
    * Remote repository သို့ပြောင်းလဲမှုများအား Upload ပြုလုပ်သည်။
    <br/><br/>
* git pull
    * Remote repository မှ ပြောင်းလဲမှုများအား မိမိကွန်ပျူတာအတွင်းသို့ Download ရယူသည်။
